```python
import pandas as pd 
import numpy as np 
```

```python
df_raw = pd.read_csv("../data/raw/Flipkart_Mobiles.csv")
```

```python
print("Raw Flipkart Phone dataset-Initial check\n")
print(f"Dataset shape: {df_raw.shape[0]} rows * {df_raw.shape[1]} coloums")
print("*"*60)
print("Coloums Names and Data types")
print(df_raw.dtypes)
print("*"*60)

print(f"\nFirst 3 rows:")
print("*"*60)

print(df_raw.head(3))
```

```python
# 1) Standardize column names
clean_cols = (
    df_raw.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df_raw.columns = clean_cols

print("Updated column names:")
print(df_raw.columns)
```

```python
for col in ["rating", "selling_price", "original_price"]:
    if col in df_raw.columns:
        df_raw[col] = pd.to_numeric(df_raw[col])

print(df_raw[["rating", "selling_price", "original_price"]].dtypes)
```

```python
#Converting memory to numeric (GB)
if "memory" in df_raw.columns:
    df_raw["memory_gb"] = (
        df_raw["memory"]
        .astype(str)
        .str.replace("GB", "", regex=False)
        .str.replace(" ", "", regex=False)
        .replace("nan", np.nan)
    )
    df_raw["memory_gb"] = pd.to_numeric(df_raw["memory_gb"], errors="coerce")

print(df_raw[["memory", "memory_gb"]].head(3))
```

```python
#storage converted into numeric
if "storage" in df_raw.columns:
    df_raw["storage_gb"] = (
        df_raw["storage"]
        .astype(str)
        .str.replace("GB", "", regex=False)
        .str.replace(" ", "", regex=False)
        .replace("nan", np.nan)
    )
    df_raw["storage_gb"] = pd.to_numeric(df_raw["storage_gb"], errors="coerce")

print(df_raw[["storage", "storage_gb"]].head(3))
```

```python
#missing values 

print("Missing values per column:")
print(df_raw.isnull().sum())
```

```python
#remving duplicated values 
print(f"Duplicate rows before removal: {df_raw.duplicated().sum()}")
df_raw = df_raw.drop_duplicates()
print(f"Duplicate rows after removal: {df_raw.duplicated().sum()}")
print(f"Final dataset shape: {df_raw.shape}")
```

```python
import os

os.makedirs("../data/cleaned", exist_ok=True)
df_raw.to_csv("../data/cleaned/flipkart_mobiles_cleaned.csv", index=False)
# print("Saved: ../data/cleaned/flipkart_mobiles_cleaned.csv")
```
