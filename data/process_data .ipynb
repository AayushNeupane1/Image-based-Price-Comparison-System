```python
import pandas as pd

df = pd.read_csv('products_all.csv')
print("Before changes:")
print(f"Columns : {list(df.columns)}")
print(f"Total rows : {len(df)}")
print(df.head(3))
```

```python
df = df.drop(columns=['in_stock'])

print("in_stock column dropped!")
print(f"Columns now : {list(df.columns)}")
```

```python
df['product_id'] = ['P' + str(i).zfill(4) for i in range(1, len(df) + 1)]

print("Product IDs updated!")
print(df['product_id'].head(10))
```

```python
df.to_csv('../products_all_fixed.csv', index=False)

print("products_all_fixed.csv saved successfully!")
```
